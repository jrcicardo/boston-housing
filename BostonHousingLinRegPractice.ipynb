```python
houses = houses.drop(columns=['ID'])
```

```python
from sklearn.linear_model import LinearRegression as LinReg
from sklearn import preprocessing 
```

```python
scaled_houses = preprocessing.scale(houses)
```

```python
scaled_houses_df= pd.DataFrame(scaled_houses, columns = houses.columns)
```

```python
scaled_houses_df = scaled_houses_df.drop(columns = ['ID'])
scaled_houses_df.head()
```

```python
scaled_y = scaled_houses_df['medv']
scaled_X = scaled_houses_df[['crim','indus','nox','rm','age','tax','ptratio','lstat']]
lr = LinReg(n_jobs=-1)
lr.fit(scaled_X,scaled_y)
scaled_y_pred = lr.predict(scaled_X)
plt.scatter(scaled_y_pred,scaled_y)
plt.plot(scaled_y,scaled_y, color = 'orange')
plt.show()
```

```python
from sklearn import metrics
metrics.explained_variance_score(scaled_y,scaled_y_pred)
```

```python
y = houses['medv']
X = houses[['crim','indus','nox','rm','age','tax','ptratio','lstat']]
lr = LinReg(n_jobs=-1)
lr.fit(X,y)
y_pred = lr.predict(X)
plt.scatter(y_pred,y)
plt.plot(y,y, color = 'orange')
plt.show()
```

```python
#the model looks...not great
#the R-squared value is less than stellar
#it looks like 10 or so towns' median house values behave a lot differently than the rest of the towns, so let's drop those
```

```python
houses = houses.sort_values('medv')
houses = houses.drop(houses.tail(11).index)
```

```python
highways = houses.loc[houses['rad']<10]
```

```python
highway_y = highways['medv']
# houses['weighted_indus'] = 2*houses['indus']
highway_X = highways[['crim','rm','tax','ptratio','lstat']]
lr = LinReg(n_jobs=-1)
lr.fit(highway_X,highway_y)
y_pred = lr.predict(highway_X)
plt.scatter(y_pred,highway_y)
plt.plot(highway_y,highway_y, color = 'orange')
plt.show()
```

```python
metrics.explained_variance_score(highway_y,y_pred)
```

```python
y = houses['medv']
# houses['weighted_indus'] = 2*houses['indus']
X = houses[['crim','rm','tax','ptratio','lstat','jobs','schools','air_quality','sad','log_dis']]
lr = LinReg(n_jobs=-1)
lr.fit(X,y)
y_pred = lr.predict(X)
plt.scatter(y_pred,y)
plt.plot(y,y, color = 'orange')
plt.show()
```

```python
metrics.explained_variance_score(y,y_pred)
```

```python
#that's a little better, but maybe we can try to make some interaction features to make the model even better
```

```python
houses['jobs'] = houses['indus']/houses['log_dis']
```

```python
houses['schools'] = houses['tax']*houses['ptratio']
```

```python
houses['air_quality'] = houses['nox']*(houses['indus'] )*houses['age']
```

```python
houses['money_money'] = houses['crim']*houses['tax']
```

```python
houses['sad'] = houses['crim']*houses['lstat']
```

```python
#The interaction features improve the model a little bit, but still leave over 20% of the variance unexplained.
#Looking at the histograms of each column, it looks like there's a cluster of high-tax-rate towns.  The data from 
#these towns may behave differently than the rest of the dataset.  Let's split the data into two groups based on tax 
#rate.
```
